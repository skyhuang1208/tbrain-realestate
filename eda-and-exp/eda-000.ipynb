#### Skewness

```python
skewness_train = df_train[cols_num].skew()
skewness_test = df_test[cols_num].skew()
```

```python
skewness_train[skewness_train > 10].sort_values()
```

```python
skewness_train[skewness_train < -1.5].sort_values()
```

#### Distribution

```python
plot_c = skewness_train.sort_values().index
for c in plot_c:
    print(c, df_train[c].skew(), df_test[c].skew())
    plt.figure(figsize=(14, 3))
    sq = df_train[c]
    sq = sq[np.isfinite(sq)]
    plt_hist_norm(sq, bins=100, label='train', alpha=1)
    sq = df_test[c]
    sq = sq[np.isfinite(sq)]
    plt_hist_norm(sq, bins=100, label='test', alpha=0.7)
    plt.legend(); plt.xlabel(c); plt.yscale('log')
    plt.show()
```

#### Right skew feature distribution

```python
plot_c = skewness_train[skewness_train > 25].sort_values().index
for c in plot_c:
    print(c, df_train[c].skew(), df_test[c].skew())
    plt.figure(figsize=(14, 3))
    sq = df_train[c]
    sq = np.log10(sq[np.isfinite(sq)]+0.1)
    plt_hist_norm(sq, bins=100, label='train', alpha=1)
    sq = df_test[c]
    sq = np.log10(sq[np.isfinite(sq)]+0.1)
    plt_hist_norm(sq, bins=100, label='test', alpha=0.7)
    plt.legend(); plt.xlabel('log10('+c+'+0.1)'); plt.yscale('log')
    plt.show()
```
